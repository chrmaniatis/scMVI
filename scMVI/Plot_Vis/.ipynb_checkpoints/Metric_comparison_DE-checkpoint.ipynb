```python
def plot_fnc(x1,x2,ttl,sv_lnk):
        y1 = np.log(x1.sort_index().to_numpy())
        y2 = np.log(x2.sort_index().to_numpy())/np.log2(np.exp(1))
        x = np.linspace(np.minimum(np.min(y1),np.min(y2)),np.maximum(np.max(y1),np.max(y2)))
        plt.scatter(y1,y2)
        plt.plot(x,x,'r')
        plt.xlabel("DDLFC Range (Log)",fontsize=15)
        plt.ylabel("LFC Range (Log)",fontsize=15)
        plt.title(ttl,fontsize=15)
        #plt.savefig(sv_lnk)
        plt.show()
        print(np.mean(y1<y2))
```

```python
#Estimate correlation between LFC and DDLFC.
def plot_lfc(x1,x2,ttl,sv_link):
      xxx = np.linspace(np.min(x1),np.max(x2),100)

      plt.scatter(x1,x2)
      plt.plot(xxx,xxx,'r')
      plt.xlabel("DDLFC",fontsize=15)
      plt.ylabel("LFC",fontsize=15)
      plt.title(ttl,fontsize=15)
      #plt.savefig(sv_link)
      plt.show()
      print(np.corrcoef(x1,x2))

lnk = Root_Folder + code_parent_folder + "/Paper_Plots/E80_E85_DDLFC_comp.png"
plot_lfc(scMVI_exp_all.loc[:,"exp1"],scVI_exp_all.loc[:,"exp6"],"E8.0 to E8.5",lnk)
lnk = Root_Folder + code_parent_folder + "/Paper_Plots/E70_E75_DDLFC_comp.png"
plot_lfc(scMVI_exp_all.loc[:,"exp4"],scVI_exp_all.loc[:,"exp3"],"E7.0 to E7.5",lnk)

lnk = Root_Folder + code_parent_folder + "/Paper_Plots/CD14vsCD4_comp.png"
plot_lfc(scMVI_exp_all_PBMC.loc[:,"exp1"],MultiVI_exp_all.loc[:,"exp1"],"CD14 vs CD4",lnk)
lnk = Root_Folder + code_parent_folder + "/Paper_Plots/NKvsMemoryB_comp.png"
plot_lfc(scMVI_exp_all_PBMC.loc[:,"exp2"],MultiVI_exp_all.loc[:,"exp2"],"NK vs MemoryB",lnk)

lnk = Root_Folder + code_parent_folder + "/Paper_Plots/CD16MonovsCD14Mono_comp.png"
plot_lfc(scMVI_exp_all_PBMC.loc[:,"exp1"],MultiVI_exp_all.loc[:,"exp1"],"CD16 Mono vs CD14 Mono",lnk)
lnk = Root_Folder + code_parent_folder + "/Paper_Plots/CD4TEMvsCD4TCM_comp.png"
plot_lfc(scMVI_exp_all_PBMC.loc[:,"exp2"],MultiVI_exp_all.loc[:,"exp2"],"CD4 TEM vs CD4 TCM",lnk)
```

```python
#Choose genomic regions with the stongest positive or negative DDLFC
def choose_lab(x,num_fts,lab):  

              ind_srt = np.argsort(x.to_numpy(),axis=0)

              for ii in range(ind_srt.shape[1]):

                    ind_lw = pd.DataFrame(x.iloc[ind_srt[:,ii][:num_fts],ii].index)
                    ind_up = pd.DataFrame(x.iloc[ind_srt[:,ii][-num_fts:],ii].index)

                    ind_lw.to_csv(Root_Folder + code_parent_folder + "/scVI_scMVI_comp/"+lab+"_lw_"+str(ii)+".csv",header=False,sep='\t',index=False)
                    ind_up.to_csv(Root_Folder + code_parent_folder + "/scVI_scMVI_comp/"+lab+"_up_"+str(ii)+".csv",header=False,sep='\t',index=False)


              ind_all = pd.DataFrame(x.index)
               
              ind_all.to_csv(Root_Folder + code_parent_folder + "/scVI_scMVI_comp/all_"+lab+".csv",header=False,sep='\t',index=False)     

choose_lab(scMVI_exp_all,500,"scMVI")
choose_lab(scVI_exp_all,500,"scVI")

choose_lab(scMVI_exp_all_NMT_acc,500,"scMVI_acc")
choose_lab(scMVI_exp_all_NMT_met,500,"scMVI_met")
choose_lab(scMVI_exp_all_NMT_rna,500,"scMVI_rna")
```
