```python
from IPython import display
import pandas as pd
import numpy as np
import datetime
from matplotlib import pyplot as plt
from matplotlib import colors
from matplotlib.ticker import PercentFormatter
from matplotlib import cm
from matplotlib.colors import Normalize 
from tensorflow import keras
import tensorflow as tf
import scipy.stats as stats
from tensorflow.keras import layers
import tensorflow_probability as tfp
from scipy.stats import gaussian_kde
import os


#Load Data
Root_Folder = "/content/drive/MyDrive" 
data = 'gastr'

data_parent_folder = "/Data"
code_parent_folder = "/scMVI"

os.chdir('/content/drive/My Drive/' + code_parent_folder +'/Code')

Met_test=pd.read_csv(Root_Folder + data_parent_folder + "/Data_Gastr/Met_test_" + data + ".csv",",")
CpG_test=pd.read_csv(Root_Folder + data_parent_folder + "/Data_Gastr/CpG_test_" + data + ".csv",",")
Acc_test=pd.read_csv(Root_Folder + data_parent_folder + "/Data_Gastr/Acc_test_" + data + ".csv",",")
GpC_test=pd.read_csv(Root_Folder + data_parent_folder + "/Data_Gastr/GpC_test_" + data + ".csv",",")
Rna_test=pd.read_csv(Root_Folder + data_parent_folder + "/Data_Gastr/Rna_test_" + data + ".csv",",")
Rna_nrm_test=pd.read_csv(Root_Folder + data_parent_folder + "/Data_Gastr/Rna_nrm_test_" + data + ".csv",",")
```

```python
MMet_test=Met_test.drop(columns=Met_test.columns[0])
MCpG_test=CpG_test.drop(columns=CpG_test.columns[0])
MAcc_test=Acc_test.drop(columns=Acc_test.columns[0])
MGpC_test=GpC_test.drop(columns=GpC_test.columns[0])
MRna_test=Rna_test.drop(columns=Rna_test.columns[0])
MRna_nrm_test=Rna_nrm_test.drop(columns=Rna_nrm_test.columns[0])

Met_tst=tf.convert_to_tensor(MMet_test,dtype=tf.float32)
CpG_tst=tf.convert_to_tensor(MCpG_test,dtype=tf.float32)
Acc_tst=tf.convert_to_tensor(MAcc_test,dtype=tf.float32)
GpC_tst=tf.convert_to_tensor(MGpC_test,dtype=tf.float32)
Rna_tst=tf.convert_to_tensor(MRna_test,dtype=tf.float32)
Rna_nrm_tst= tf.math.log(0.000001+tf.convert_to_tensor(MRna_nrm_test,dtype=tf.float32))

inp_met_sz = Met_tst.shape[1]
inp_acc_sz = Acc_tst.shape[1]
inp_rna_sz = Rna_tst.shape[1]
```

```python
from Models import scMVI_NMT
from Obj_Blocks import *
from Plotting import *
```

```python
#Define model and load weights for the trained model
mdl = scMVI_NMT(inp_met_dim = inp_met_sz,inp_acc_dim=inp_acc_sz,inp_rna_dim=inp_rna_sz,intermidiate_dim=100,latent_dimensions=10)
mdl.load_weights(Root_Folder + code_parent_folder + "/Networks/scMVI_scNMT_"+data+"/scMVI_scNMT_"+data)
```

```python
#Estimate uncertainty for reconstruced or imputed samples
def est_std(x):
    x = tf.stack(x,axis=0)
    x = tfp.stats.stddev(x,sample_axis=0)
    return x


def data_uncertainty(model,xx,num_samps):
    col_smps = []
    col_imp_smps = []
    for ii in range(num_samps):

          Met_pred,Acc_met_pred,Rna_met_pred = mdl.imp_cells([Met_tst,CpG_tst,Acc_tst,GpC_tst,Rna_tst,Rna_nrm_tst],'met')
          Met_acc_pred,Acc_pred,Rna_acc_pred = mdl.imp_cells([Met_tst,CpG_tst,Acc_tst,GpC_tst,Rna_tst,Rna_nrm_tst],'acc')
          Met_rna_pred,Acc_rna_pred,Rna_pred = mdl.imp_cells([Met_tst,CpG_tst,Acc_tst,GpC_tst,Rna_tst,Rna_nrm_tst],'rna')

          col_smps.append([Met_pred,Acc_pred,tf.math.log(1+Rna_pred)])
          col_imp_smps.append([Met_rna_pred,Met_acc_pred,Acc_rna_pred,Acc_met_pred,tf.math.log(1+Rna_met_pred),tf.math.log(1+Rna_acc_pred)])
    
    
    smps_std = est_std(col_smps)
    imp_smps_std = est_std(col_imp_smps)

    return col_smps,smps_std,col_imp_smps,imp_smps_std

def find_ind(CpG_dat,GpC_dat,Rna_dat):
    avg_cov_met=tf.reduce_mean(CpG_dat,axis=1)
    avg_cov_acc=tf.reduce_mean(GpC_dat,axis=1)
    avg_exp=tf.reduce_mean(Rna_dat,axis=1)

    cpg_ind_zer = tf.where(avg_cov_met==0)
    gpc_ind_zer = tf.where(avg_cov_acc==0)
    rna_ind_zer = tf.where(avg_exp==0)

    cpg_ind_nzer = tf.where(avg_cov_met>0)
    gpc_ind_nzer = tf.where(avg_cov_acc>0)
    rna_ind_nzer = tf.where(avg_exp>0)

    cells_met_mod = np.intersect1d(cpg_ind_nzer,rna_ind_zer)
    cells_met_mod = np.intersect1d(cells_met_mod,gpc_ind_zer)

    cells_acc_mod = np.intersect1d(gpc_ind_nzer,rna_ind_zer)
    cells_acc_mod = np.intersect1d(cells_acc_mod,cpg_ind_zer)

    cells_rna_mod = np.intersect1d(cpg_ind_zer,rna_ind_nzer)
    cells_rna_mod = np.intersect1d(cells_rna_mod,gpc_ind_zer)

    cells_jt_mod = np.intersect1d(cpg_ind_nzer,rna_ind_nzer)
    cells_jt_mod = np.intersect1d(cells_jt_mod,gpc_ind_nzer)

    return cells_met_mod,cells_acc_mod,cells_rna_mod,cells_jt_mod
```

```python
smps,smps_std,smps_imp,smps_imp_std = data_uncertainty(mdl,[Met_tst,CpG_tst,Acc_tst,GpC_tst,Rna_tst,Rna_nrm_tst],20)
#Met_pred_tst,Acc_pred_tst,Rna_pred_tst = smps
#Met_rna_pred_tst,Met_acc_pred_tst,Acc_rna_pred_tst,Acc_met_pred_tst,Rna_met_pred_tst,Rna_acc_pred_tst = smps_imp

#Met_pred_tst_std = tf.squeeze(smps_std[0,:,:])
#Acc_pred_tst_std = tf.squeeze(smps_std[1,:,:])
#Rna_pred_tst_std = tf.squeeze(smps_std[2,:,:])

#Met_rna_pred_std = tf.squeeze(smps_imp_std[0,:,:])
Met_acc_pred_std = tf.squeeze(smps_imp_std[1,:,:])
#Acc_rna_pred_std = tf.squeeze(smps_imp_std[2,:,:])
#Acc_met_pred_std = tf.squeeze(smps_imp_std[3,:,:])
#Rna_met_pred_std = tf.squeeze(smps_imp_std[4,:,:])
Rna_acc_pred_std = tf.squeeze(smps_imp_std[5,:,:]) 
```

```python
Rna_acc_pred_tst = []
Met_acc_pred_tst = []
for ii in range(len(smps_imp)):
      Rna_acc_pred_tst.append(smps_imp[ii][-1])
      Met_acc_pred_tst.append(smps_imp[ii][1])

Rna_acc_pred_tst = tf.stack(Rna_acc_pred_tst,axis=0)
Met_acc_pred_tst = tf.stack(Met_acc_pred_tst,axis=0)
```

```python
#dist = tf.math.sqrt(tf.reduce_mean(tf.math.square(Rna_acc_pred_tst-tf.math.log(1+Rna_tst)),axis=0))
dist_rna = tf.reduce_mean(tf.math.abs(Rna_acc_pred_tst-tf.math.log(1+Rna_tst)),axis=0)
dist_met = tf.reduce_mean(tf.math.abs(Met_acc_pred_tst-Met_tst),axis=0)
std_rna = Rna_acc_pred_std
std_met = Met_acc_pred_std

_,_ ,_,ind_jt_tst = find_ind(CpG_tst,GpC_tst,Rna_tst)
```

```python
from matplotlib import pyplot as plt
from matplotlib import colors
from matplotlib.ticker import PercentFormatter
from matplotlib import cm
from matplotlib.colors import Normalize 
import scipy.stats as stats
from scipy.stats import gaussian_kde
import numpy as np
import tensorflow as tf

def density_scatter_plot_1(x,y, **kwargs):
    """
    :param x: data positions on the x axis
    :param y: data positions on the y axis
    :return: matplotlib.collections.PathCollection object
    """
    # Kernel Density Estimate (KDE)
    values = np.vstack((x, y))
    kernel = gaussian_kde(values)
    kde = kernel.evaluate(values)

    # create array with colors for each data point
    norm = Normalize(vmin=kde.min(), vmax=kde.max())
    colors = cm.ScalarMappable(norm=norm, cmap='viridis').to_rgba(kde)

    # override original color argument
    kwargs['color'] = colors


    plt.scatter(x, y, **kwargs)
    plt.colorbar(orientation='vertical')
    
    return plt
```

```python
aaa = ind_jt_tst
col_prob_rna = np.zeros(10,)
col_prob_met = np.zeros(10,)
for ii in range(10):
          ii+1
          xx_rna = tf.math.log(0.001+2.0*std_rna[aaa[ii],:])
          yy_rna = tf.math.log(0.001+dist_rna[aaa[ii],:])

          xx_met = tf.math.log(0.001+2.0*std_met[aaa[ii],:])
          yy_met = tf.math.log(0.001+dist_met[aaa[ii],:])

          xx1 = tf.linspace(tf.reduce_min(xx_rna),tf.reduce_max(xx_rna),100)
          xx2 = tf.linspace(tf.reduce_min(xx_met),tf.reduce_max(xx_met),100)

          plt=density_scatter_plot_1(xx_rna,yy_rna) 
          plt.plot(xx1,xx1,'r')
          plt.xlabel("Confidense Interval Length (log)",fontsize=15)
          plt.ylabel("Imputation Error (log)",fontsize=15)
          #plt.savefig(Root_Folder + code_parent_folder +"Paper_Plots/NMT_imput_diagram.png")
          plt.show()

          plt=density_scatter_plot_1(xx_met,yy_met) 
          plt.plot(xx2,xx2,'r')
          plt.show()

          problematic_rna = tf.where(yy_rna>xx_rna)
          problematic_met = tf.where(yy_met>xx_met)
          col_prob_rna[ii] = problematic_rna.numpy().shape[0]
          col_prob_met[ii] = problematic_met.numpy().shape[0]
```
